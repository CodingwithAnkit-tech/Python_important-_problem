```python
"""1.Given a number, write a nested if program to check: 
If it’s divisible by 2 Then check if it’s divisible by 4 If yes → print "Divisible by 4" 
Else → print "Even but not divisible by 4" Else → print "Odd number" """

def is_even_odd():
    n = int(input("Enter the number :"))
    if n % 2 == 0:
        if n % 4 == 0:
            print("Divisible by 4")
        else:
            print("Even but not divisible by 4")
    else:
        print("Odd number")
is_even_odd()

```

```python
"""2.ATM Cash Withdrawal Simulate an ATM: Input balance and amount to withdraw. 
If amount ≤ balance If amount is a multiple of 100 → print "Transaction Successful" 
Else → "Enter amount in multiples of 100" Else → "Insufficient Balance" """

def ATM_Withdrawal():
    input_balance = int(input("Enter the input balance :"))
    amount_withdraw = int(input("Enter the withdrawal :"))
    if amount_withdraw <= input_balance:
        if amount_withdraw % 100 == 0:
            print("Transaction Successfull")
        else:
            print("Enter amount in multiples of 100")
    else:
        print("Insufficient Balance")


ATM_Withdrawal()
```

```python
def ATM_Withdrawal():
    input_balance = int(input("Enter the input balance : "))
    amount_withdraw = int(input("Enter the withdrawal : "))

    if amount_withdraw > input_balance:
        return "Insufficient Balance"

    if amount_withdraw % 100 != 0:
        return "Enter amount in multiples of 100"

    return "Transaction Successful"


result = ATM_Withdrawal()
print(result)
```

```python
# 3.Wap to check the given data is pallindrome or not only if the given data is string.

def check_palindrome():
    data = input("Enter data: ")

    if isinstance(data, str):
        if data == data[::-1]:
            print("Palindrome string")
        else:
            print("Not a palindrome")
    else:
        print("Enter string")

check_palindrome()


```

```python
# 5.Wap to check the given data is vowel or consonant.

def vowel_consonant():
    ch = input("Enter character: ").lower()

    if ch.isalpha():
        if ch in "aeiou":
            print("Vowel")
        else:
            print("Consonant")
    else:
        print("Not an alphabet")

vowel_consonant()
```

```python
# 6 Wap to find the smallest no. among 3 numbers by using nested if

def smallest_number():
    a = int(input("Enter first number: "))
    b = int(input("Enter second number: "))
    c = int(input("Enter third number: "))

    if a < b:
        if a < c:
            print("Smallest:", a)
        else:
            print("Smallest:", c)
    else:
        if b < c:
            print("Smallest:", b)
        else:
            print("Smallest:", c)

smallest_number()
```
