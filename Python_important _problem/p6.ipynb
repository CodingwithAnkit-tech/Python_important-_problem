```python
# 8.Wap to print n natural palindrome numbers

def print_palindrome_numbers():
    n = int(input("Enter value of n: "))
    count = 0
    num = 1

    while count < n:
        if str(num) == str(num)[::-1]:
            print(num)
            count += 1
        num += 1

# function call
print_palindrome_numbers()

```
