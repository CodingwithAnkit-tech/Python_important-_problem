```python
#2. Factorial of a Number
# Ans. Write a function to calculate the factorial of a nonnegative integer.

def factorial(n):
    if n == 0:
        return 1
    return n* factorial(n-1)

print(factorial(4))   # 5 ka factorial print karega → 24
```

```python
#3. Problem: Fibonacci Sequence
#Ans . Write a function to generate the nth number in the Fibonacci sequence

def fibonacci(n):
    if n<=1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)
# print(fibonacci(6)) # output 8
# user input

n = int(input("Enter n: "))
print(fibonacci(n))


```

```python
# 4. Problem: Palindrome Check
#Ans. Write a function to check if a given string is a palindrome

def is_palindrome(s):
    s = str(s).lower().replace("","")
    return s == s[::-1]
print(is_palindrome("madam"))
print(is_palindrome("hello"))
print(is_palindrome(121))
```

```python
#5. Problem: Prime Number Check
#Ans. Write a function to check if a given number is prime


def is_prime(n):
    if n<=1:
        return False
    for i in range(2,int(n**0.5)+ 1):
        if n% i == 0:
            return False
    return True

print(is_prime(7))   # True
print(is_prime(12))  # False


```

```python
# 6. Problem: Reverse a String
# Ans. Write a function to reverse a given string.

def reverse_string(s):
    return s[::-1]
print(reverse_string("aman"))
```

```python
# 7. Problem: Count Digits in a Number
# Ans. Write a function to count the number of digits in a positive integer

def count_digit(n):
    return len(str(n))
print(count_digit("raman"))
```

```python
# 8. Problem: Find Maximum Number
# Ans. Write a function to check if a given integer is even or odd

def is_even_or_odd(n):
    if n % 2 == 0:
        return "Even"
    else:
        return "Odd"

print(is_even_or_odd(45))  # Output: Odd
print(is_even_or_odd(40))  # Output: Even

```

```python
# wap to print amstrong number
def is_armstrong_number(n):
    s = str(n)
    num_digits = len(s)
    armstrong_sum = sum(int(digit) ** num_digits for digit in s)
    return armstrong_sum == n
n = int(input("Enter a number: "))
if is_armstrong_number(n):
    print(f"{n} is an Armstrong number.")
else:
    print(f"{n} is not an Armstrong number.")
    
```

```python
# extract all the upper case lower case digit and special character from the given string
upp =''
low =''
dig =''
spc =''
def extract(s):
    for i in s:
        if i.isupper():
            global upp
            upp += i
        elif i.islower():
            global low
            low += i
        elif i.isdigit():
            global dig
            dig += i
        else:
            global spc
            spc += i
    print("upper case:",upp)
    print("lower case:",low)
    print("digit:",dig)
    print("special character:",spc)
extract("Hello@123")
```
